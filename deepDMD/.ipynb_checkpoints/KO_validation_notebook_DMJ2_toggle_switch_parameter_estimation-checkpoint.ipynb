```python
# Number of time points we want for the solutions
n = 101

# Time points we want for the solution
t = np.linspace(0, 20, n)

# Parameters
beta_y = 5.0
beta_z = 5.0
n_x = 2
n_y = 3
x_0 = 2.0
Km1 = 1
Km2 = 2
delta = 0.18

# Package parameters into a tuple
args = (beta_y, beta_z, n_x, n_y, Km1, Km2, delta)

def toggle_switch(yz, t, beta_y, beta_z, n_x, n_y, Km1, Km2, delta):
    """
    Right hand side for cascade X -> Y -> Z.  Return dy/dt and dz/dt.
    """
    # Unpack y and z
    y, z = yz
    
    # Compute dy/dt
    dy_dt = beta_y/ (1 + (z/Km1)**n_x) - delta*y
    
    # Compute dz/dt
    dz_dt = beta_z/ (1 + (y/Km2)**n_y) - delta*z
    
    # Return the result as a NumPy array
    return np.array([dy_dt, dz_dt])
```

# Hill function fitting

```python
x1 = np.linspace(0, 20, n)
x2 = np.linspace(0, 20, n)

#x_c = np.linspace(-10, 10, n)
#y_c = np.linspace(-10, 10, n)

X, Y = np.meshgrid(x1, x2)

#psixpT.eval(feed_dict={xpT_feed: x_learned}).T[2]

def repressor_function1(y, z, alpha_y, alpha_z, beta_y, beta_z, n_y, n_z, Km1, Km2, Vmax):
    return Vmax*(1 + beta_y*(z/Km1)**n_z + beta_z*(y/Km2)**n_y)/(1 + alpha_z*(z/Km1)**n_z + alpha_y*(y/Km2)**n_y)

# A list of the Hill function parameters: x0, y0, xalpha, yalpha, A
hf_prms = [(1, 1, 1, 0, 2, 2, 12, 12, 10)]

# The function to be fit is Z.
Z = np.zeros(X.shape)
for p in hf_prms:
    Z += repressor_function1(X, Y, *p)
```

```python
# Plot the 3D figure of the fitted function and the residuals.
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(X, Y, Z, cmap='plasma')
ax.set_zlim(0,np.max(Z)+2)
plt.show()
```

```python
# This is the callable that is passed to curve_fit. M is a (2,N) array
# where N is the total number of data points in Z, which will be ravelled
# to one dimension.
def _repressor_function1(M, *args):
    x, y = M
    arr = np.zeros(x.shape)
    #print(args)
    for i in range(len(args)//9):
        arr += repressor_function1(x, y, *args[i*9:i*9+9])
    return arr
```

```python
# We need to ravel the meshgrids of X, Y points to a pair of 1-D arrays.
xdata = np.vstack((X.ravel(), Y.ravel()))
# Do the fit, using our custom _gaussian function which understands our
# flattened (ravelled) ordering of the data points.
```

```python

learned_params = []
error_thresh = 0.0001
fit_error = 100
rms_list = []
# Flatten the initial guess parameter list.
while fit_error > error_thresh:
    #print(simulation)
    p0 =  [(np.random.randint(0, 3), np.random.randint(0, 3), np.random.randint(5, 20), np.random.randint(5, 20), np.random.randint(5, 20), np.random.randint(5, 20), np.random.randint(5, 20), np.random.randint(5, 20), np.random.randint(5, 20))]

    popt, pcov = curve_fit(_repressor_function1, xdata, scaled_Psi_obs1.ravel(), p0[0], bounds=(0, np.inf), maxfev = 5000);
    fit = np.zeros(scaled_Psi_obs1.shape);
    #fit = np.zeros(Z.shape)
    for i in range(len(popt)//9):
        fit += repressor_function1(X, Y, *popt[i*9:i*9+9])
    fit_error = np.sqrt(np.mean((scaled_Psi_obs1 - fit)**2))
    print(fit_error)
    learned_params.append(popt)
    rms_list.append(fit_error)
```
