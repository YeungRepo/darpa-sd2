```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#trajDir = './Trajectories'
trajDir = '/Volumes/Projects-1/DARPA_SD2/ToggleSSA/ToggleTrajectories' #Path on Rob's machine
```

```python
def find_nearest(array,value):
    idx = (np.abs(array-value)).idxmin()
    return idx

def extractAbundance(trajDirectory,trajNum,timeSamples):

    df = pd.read_csv(trajDirectory+'/Toggle_traj{:05d}.csv'.format(trajNum),index_col=0)
    idx = []
    for i in timeSamples:
        idx.append( find_nearest(df.Time,i) )
    #print('Number of time points: {}'.format(len(idx)))
    dfT = df[['LacI','TetR']].iloc[idx]
    return dfT.as_matrix()
```

```python
maxTime = 250
timeSamples = np.arange(0,maxTime,0.2)

vals = np.zeros((10000,len(timeSamples),2))
# 10,000 trajectories 
# x time samples
# LacI & TetR counts

for traj in np.arange(1,10000):
#for traj in np.arange(1,5):
    temp = extractAbundance(trajDir, traj, timeSamples)
    vals[traj] = temp
```

```python
np.save(trajDir+'/../toggleAry',vals)
```
