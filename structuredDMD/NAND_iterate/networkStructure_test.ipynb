```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Creating a Graph 
G = nx.Graph() # empty undirected graph

DiG = nx.DiGraph() # empty directed graph
```

```python
# Add a node
G.add_node(1) 
G.add_nodes_from([2,3]) # You can also add a list of nodes by passing a list argument

DiG.add_node(1) 
DiG.add_nodes_from([2,3]) # You can also add a list of nodes by passing a list argument
```

```python
# Add edges 
G.add_edge(1,2)

DiG.add_edge(1,2)
```

```python
e = (2,3)
G.add_edge(*e) # * unpacks the tuple
G.add_edges_from([(1,2), (1,3)]) # Just like nodes we can add edges from a list

DiG.add_edge(*e) # * unpacks the tuple
DiG.add_edges_from([(1,2), (1,3)]) # Just like nodes we can add edges from a list
```
